```python
# Problem 1(c(i))
#adding an extra column 'Average Room Size'
dataset['Average Room Size'] = dataset['Living Space'] / dataset['Bedrooms Count'] 
```

```python
# Problem 1(c(ii))
#adding an extra column '% of Area under Living space'
dataset['% of Area under Living space'] = dataset['Living Space'] / dataset['Area_Sqft'] 
```

```python
# To check the 1st five row and feature variables
dataset.head()
```

```python
# Problem 1(d)
# for building a multiple regression model we should have to select the independent and dependent variables
X = dataset.iloc[:, [1,2,3,4,5,6,7,8]].values   # independent variables
y = dataset.iloc[:, 11].values                  # dependent variables
```

```python
# Build a Machine Learning Modelto predict “Selling Price” through other variables
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2, random_state = 0) # we take 80% tranning set and 20% test dataset
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# Import the sklearn libraries
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)    # Coefficient of determination
```

```python
# The coefficient of determination is 0.8550047322613521 means that 85 percent of the variance in Y is predictable from X.
```

```python
# Building a model (Backword Elimination)
# Using Backword Elimination we will select the actual independent variable to make the model accurate.
# By observing p-value(0.05) we will decide whitch wanna keep (independent variable) and whitch wanna through out. 
```

```python
# Backword Elemination step by step
#step-1 select a significance level to stay in the model (e.g Sl=0.05)
#step-2 Fill the full model with all possible predictor
#step-3 Consider the predictor with the hightest p-value, if P > Sl=0.05, go to next step otherwise finish
#step-4 Remove the predictor
#step-5 Fit the model without this variable
```

```python
# Building a optimal model (For the multiple linear regression model we will add an extra column for the the constant value(b0) whitch is 57*1 vector)
import statsmodels.api as sm
X = np.append(arr = np.ones((57, 1)).astype(int), values = X, axis = 1)
```

```python
# Using Ordinary Least Square Method we will calculate p-value for all independent variables
X_opt = X[:, [0, 1, 2, 3, 4, 5,6,7,8]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit() # X_opt is the dependent variable and y is the dependet variable.
regressor_OLS.summary()
```

```python
# From this result x3 = Area_Sqft has a hight p-value(0.845), Using backword elemination we will delect the following ind variable
X_opt = X[:, [0, 1, 2, 4, 5, 6, 7, 8]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# Again From the previous result x5 = Rooms Count has a hight p-value(0.826), Using backword elemination we will delect the following ind variable
X_opt = X[:, [0, 1,2, 4, 5,7,8]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# Again From this result x6 = Age has a hight p-value(0.478), Using backword elemination we will delect the following ind variable
X_opt = X[:, [0, 1,2, 4, 5,7]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# Again From this result x4 = Garrage Count has a hight p-value(0.228), Using backword elemination we will delect the following ind variable
X_opt = X[:, [0, 1,2, 4, 7]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
#Again From this result x2 = Bathrooms Count has a hight p-value(0.081), Using backword elemination we will delect the following ind variable
X_opt = X[:, [0, 1, 4, 7]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# Again From the previous result Constant Value has a hight p-value(0.081), Using backword elemination we will delect the following ind variable
X_opt = X[:, [1, 4, 7]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```
